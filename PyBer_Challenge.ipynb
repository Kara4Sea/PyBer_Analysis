### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
#import numpy as np
#import statistics as sts

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# 1. Get the total rides for each city type
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
urban_ride_count.sum()
```

```python
suburban_ride_count.sum()
```

```python
rural_ride_count.sum()
```

```python
#the type of city as the index, then apply the count() method to the "ride_id" column.
#total rides per city type

count_ride_df = pyber_data_df.groupby('type').ride_id.count()
count_ride_df
```

```python
# 2. Get the total drivers for each city type

sum_drivers_df = city_data_df.groupby('type').driver_count.sum()
sum_drivers_df

```

```python
#  3. Get the total amount of fares for each city type

total_fare_df = pyber_data_df.groupby('type').fare.sum()
total_fare_df

```

```python
#  4. Get the average fare per ride for each city type. 

avg_fare_df = pyber_data_df.groupby('type').fare.mean()
avg_fare_df


```

```python
#loc_citites_df = pyber_data_df.loc[pyber_data_df["city"] == "Lake Jonathanshire"]
#loc_citites_df
#loc_cities = pyber_data_df.loc[:,["Rodneyfort"]]
```

```python
# 5. Get the average fare per driver for each city type. 
#total fare by total driver per city type
avg_fare_driver =  total_fare_df / sum_drivers_df
avg_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
#pyber_summary_df = pyber_data_df.loc[:,["type", "ride_id", "driver_count", "fare"]]

pyber_summary_df = pd.concat([count_ride_df, sum_drivers_df, total_fare_df, avg_fare_df, avg_fare_driver], axis=1, join='inner')
pyber_summary_df.columns = ["Total Rides", "Total Drivers", "Total Fares", "Average Fare per Ride", "Average Fare per Driver"]
display(pyber_summary_df)


```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns. 

pyber_summary_df.loc[:, "Total Rides"] = pyber_summary_df["Total Rides"].map('{:,d}'.format)
pyber_summary_df.loc[:, "Total Drivers"] = pyber_summary_df["Total Drivers"].map('{:,d}'.format)
pyber_summary_df.loc[:, "Total Fares"] = pyber_summary_df["Total Fares"].map('${:.2f}'.format)
pyber_summary_df.loc[:, "Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map('${:.2f}'.format)
pyber_summary_df.loc[:, "Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map('${:.2f}'.format)
pyber_summary_df

```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
#total_fare_date_df = pyber_data_df.groupby('date').fare.sum()
#total_fare_date_df


total_fare_type_df = pyber_data_df.groupby(['type', 'date']).sum()[["fare"]]
total_fare_type_df



```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_fare_type_df = total_fare_type_df.reset_index()
total_fare_type_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_date_type_table = pyber_data_df.pivot(index='date',columns='type',values='fare')  
fare_date_type_table
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

fare_date_type_filter = fare_date_type_table.loc['2019-01-01':'2019-04-28']
fare_date_type_filter
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fare_date_type_filter.index = pd.to_datetime(fare_date_type_filter.index)
fare_date_type_filter
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_date_type_filter.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_df = fare_date_type_filter.resample('W').sum()
weekly_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

weekly_df.plot(figsize=(20, 5))
#set up x, y, title
plt.title("Total Fare by City Type", fontsize=20)
plt.ylabel("Fare ($USD)", fontsize=12)
plt.xlabel("")
#save as PyBer_fare_summary.png in analysis folder
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
